```python
import numpy as np
import pandas as pd
```

#Importing Dataset

```python
df = pd.read_csv(r'Downloads\indian movies.csv')
df
```

```python
df.info()
```

#Removing Duplicate Rows

```python
df1 = pd.DataFrame(df)
```

```python
df1.drop_duplicates(subset='ID',inplace=True, ignore_index=True, keep=False)
```

```python
df1
```

#Cleaning the Data

```python
df1['Timing(min)'] = df1['Timing(min)'].str.extract('(\d+)')
```

```python
df1.head()
```

```python
df1['Votes'] = df1['Votes'].replace('-', np.nan)  # Replace '-' with NaN
df1['Votes'] = df1['Votes'].str.replace(',', '')  # Remove commas
df1['Votes'] = pd.to_numeric(df1['Votes'], errors='coerce')  # Convert to numeric, coerce errors to NaN
```

```python
df1.head()
```

```python
df1.head(100)
```

```python
df1['Year']=df1['Year'].str.replace(r'\W', '', regex=True)
```

```python
df1['Year']= df1['Year'].str.extract('(\d+)')
```

```python
df1['ID'] = df1['ID'].str.extract('(\d+)')
```

```python
df1.head()
```

#Coverting Datatype into String

```python
df1['Genre'] = df1['Genre'].astype(pd.StringDtype())
```

```python
df1['Genre'].info()
```

#Split the multiple cell values to rows

```python
df2 = []
```

```python
a_df1 = pd.DataFrame(df1)
```

```python
df2 = (a_df1.set_index([col for col in a_df1.columns if col != 'Genre'])
               .Genre.str.split(', ', expand=True)
               .stack()
               .reset_index()
               .rename(columns={0: 'Genre'})
               .loc[:, a_df1.columns])
```

#Replacing the empty columns with null values

```python
df3 = pd.DataFrame(df2)
```

```python
df3.replace('-', pd.NA, inplace=True)
```

```python
df3
```

#Handling Missing Values

```python
df3['Timing(min)'] = df3['Timing(min)'].fillna(np.mean(pd.to_numeric(df3['Timing(min)'])))
```

```python
df1.head()
```

```python
df3['Rating(10)'] = df3['Rating(10)'].fillna(np.mean(pd.to_numeric(df3['Rating(10)'])))
```

```python
df3.head(15)
```

```python
df3['Votes'] = df3['Votes'].fillna(np.mean(pd.to_numeric(df3['Votes'])))
```

```python
df3.tail()
```

```python
df3['Year'] = df3['Year'].fillna(np.mean(pd.to_numeric(df3['Year'])))
```

```python
df3.head(20)
```

```python
df3.info()
```

```python
df3['Genre'].isnull().sum()
```

```python
df3['Genre'] = df3['Genre'].fillna(df3['Genre'].mode()[0])
```

#Converting the numerical column into int Datatypes

```python
df4 = pd.DataFrame(df3)
```

```python
df4['ID'] = df4['ID'].astype(int)
```

```python
df4['Rating(10)'] = df4['Rating(10)'].astype(float)
```

```python
df4['Timing(min)'] = df4['Timing(min)'].astype(float)
```

```python
df4.info()
```

```python
df4[(df4['Rating(10)']>9)]
```

#Importing Seaborn Package

```python
import seaborn as sns
```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(31, 11))
sns.barplot(data=df4, x=df4['Year'], y=df4['Rating(10)'])
plt.title('Distribution of movie release date')
plt.xlabel('year')
plt.ylabel('Rating')
```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(18, 11))
sns.barplot(x=df4['Genre'], y=df4['Timing(min)'])
plt.title('Distribution of Votes based on the movie gesture')
plt.xlabel('Category of movie')
plt.ylabel('Timing of movie')
```

```python
# Convert 'Year' column to datetime format
df3['Year'] = pd.to_datetime(df3['Year'],errors='coerce', format='%Y')

# Filter movies released after the year 2000 and with a rating above 7.5
filtered_movies = df4[(df4['Year'].dt.year > 2000) & (df4['Rating(10)'] > 7.5)]

print(filtered_movies)
```

```python
# Set Seaborn style
sns.set_style("whitegrid")

# Create a scatter plot of release year vs. rating
plt.figure(figsize=(19,16))
sns.scatterplot(x='Year', y='Rating(10)', data=filtered_movies ,hue='Genre', palette='Set2')

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Set plot title and labels
plt.title('Movies Released After 2000 with Rating Above 7.5')
plt.xlabel('Release Date')
plt.ylabel('Rating')

# Display the plot
plt.tight_layout()
plt.show()
```
